```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datadir = "../data"
os.listdir(datadir)
```

```python
single_mutants = pd.read_csv(os.path.join(datadir, "single_muts_train.csv"))
multi_mutants = pd.read_csv(os.path.join(datadir, "multiple_muts_train.csv"))
```

```python
single_mutants['wt_name'] = single_mutants.name.apply(lambda x: x.split(".")[0])
single_mutants['length'] = single_mutants.sequence.apply(len)
single_mutants['topology'] = single_mutants.name.apply(lambda x: x.split("_")[0])

multi_mutants['wt_name'] = multi_mutants.name.apply(lambda x: x.split(".")[0])
multi_mutants['length'] = multi_mutants.sequence.apply(len)
multi_mutants['topology'] = multi_mutants.name.apply(lambda x: x.split("_")[0])
```

## Single mutations - easier

```python
single_mutants.groupby("wt_name").mean()
```

```python
single_mutants.groupby(["wt_name", "secondary_structure"]).count()#.shape
# yes, the secondary structure is the same
```

```python
secondary_structures = single_mutants.topology.drop_duplicates().values
fig, axes = plt.subplots(ncols=len(secondary_structures)+1, figsize=(16, 4))
axes[0].hist(single_mutants.stabilityscore, bins=50)
axes[0].set_title("all")
for st, ax in zip(secondary_structures, axes[1:]):
    selection = single_mutants[single_mutants['topology'] == st]
    ax.hist(single_mutants.stabilityscore, bins=50)
    ax.hist(selection.stabilityscore, bins=50)
    ax.set_title(st)
    # ax.set_aspect("equal")
fig.suptitle("Single mutants stabilityscore distributions");
```
